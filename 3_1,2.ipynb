```python
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
g= pd.read_excel("3.xlsx", sheet_name='q')
gn= pd.read_excel("3.xlsx", sheet_name='u')
```

```python
g.head()
```

```python
gn.head()
```

```python
g["U"]= gn["U"]
g
```

```python
g["lambda"]= 0.3/g["Frequency"]
g["U/Q"]= g["U"]/g["Q"]
g
```

```python
gna= pd.read_excel("3.xlsx", sheet_name='Sheet1')
```

```python
g["arctan U/Q"]= gna["arctan U/Q"]
g["polarisation angle"]= g["arctan U/Q"]*0.5
g["wavelength ^2"]= g["lambda"]*g["lambda"]
g.head()
```

```python
plt.scatter(g["polarisation angle"], g["wavelength ^2"])
plt.gca().invert_yaxis()
plt.xlabel('polorization angle')
plt.ylabel('lambda ^2')
plt.title("Polarization angle vs lambda ^2")

plt.show()
```

```python
plt.plot(g["polarisation angle"], g["wavelength ^2"])
plt.gca().invert_yaxis()
plt.xlabel('polorization angle')
plt.ylabel('lambda ^2')
plt.title("Polarization angle vs lambda ^2")
```

```python
plt.plot(g["wavelength ^2"], g["polarisation angle"])
plt.ylabel('polorization angle')
plt.xlabel('lambda ^2')
plt.title("lambda ^2 vs polorization angle ")
```

Unwrapping the angles

```python
g["P_1"]= np.unwrap(g["polarisation angle"])
g
```

```python
plt.scatter(g["P_1"], g["wavelength ^2"])
plt.gca().invert_yaxis()
plt.xlabel('unwrapped polorization angle')
plt.ylabel('lambda ^2')
plt.title("Unwrapped Polarization angle vs lambda ^2")
plt.xlim(0.25, 0.85)
plt.ylim(0.013)
```

```python
plt.scatter(g["wavelength ^2"], g["P_1"])
plt.gca().invert_yaxis()
plt.ylabel('unwrapped polorization angle')
plt.xlabel('lambda ^2')
plt.title("lambda ^2 vs Unwrapped Polarization angle")
```

```python
plt.scatter(g["wavelength ^2"], g["P_1"])
plt.gca().invert_yaxis()
plt.ylabel('unwrapped polorization angle')
plt.xlabel('lambda ^2')
plt.title("lambda ^2 vs Unwrapped Polarization angle")
plt.ylim(0.3,0.8)
plt.xlim(0.010, 0.0135)
```

```python
plt.scatter(g["wavelength ^2"], g["P_1"])
plt.gca().invert_yaxis()
plt.ylabel('unwrapped polorization angle')
plt.xlabel('lambda ^2')
plt.title("lambda ^2 vs Unwrapped Polarization angle")
#plt.x.lim(0.013, 0.14)
plt.ylim(-0.8,0)
plt.xlim(0.012, 0.016)
```
